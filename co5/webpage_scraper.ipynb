```python
from bs4 import BeautifulSoup
import csv
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
webpage=requests.get("https://www.flipkart.com/search?q=mobile+phone&as=on&as-show=on&otracker=AS_Query_OrganicAutoSuggest_5_3_na_na_na&otracker1=AS_Query_OrganicAutoSuggest_5_3_na_na_na&as-pos=5&as-type=RECENT&suggestionId=mobile+phone&requestId=789fad67-092c-43dd-822f-68c82be5d770&as-searchtext=mob")
```

```python
webpage
```

```python
soup=BeautifulSoup(webpage.content,"html.parser")
```

```python
print(soup.prettify())
```

```python
Products=soup.find_all('div',class_='_4rR01T')
```

```python
Products
```

```python
Price=soup.find_all('div',class_='_30jeq3 _1_WHN1')
```

```python
Price
```

```python
prod_names=[]
for i in range(0,len(Products)):
  prod_names.append(Products[i].get_text())
```

```python
prod_names
```

```python
price_values=[]
for i in range(0,len(Price)):
  price_values.append(Price[i].get_text())
```
