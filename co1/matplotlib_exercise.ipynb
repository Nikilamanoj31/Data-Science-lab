```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
fig=plt.figure()

axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.2,0.5,0.2,0.2])

axes1.plot(x,y,'g')
axes1.set_xlabel('x label axes 1')
axes1.set_ylabel('y label axes 1')
axes1.set_title('title 1')

axes2.plot(x,y,'b')
axes2.set_xlabel('x label axes 2')
axes2.set_ylabel('y label axes 2')
axes2.set_title('title 2')
```

Exercise 3

Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]

```python
fig = plt.figure()

ax= fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
```

```python
ax.plot(x,z,'b')
ax.set_xlabel('X')
ax.set_ylabel('Z')


ax2.plot(x,y,'b')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)
fig
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(12,2))

axes[0].plot(x,y,color="yellow", lw=5)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

axes[1].plot(x,z,color="green", lw=3, ls='--')
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
```
